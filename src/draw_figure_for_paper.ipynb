```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_excel("D:/hyr69/문서/논문/My paper/실험결과.xlsx",sheet_name='Sheet1')
df.head()
```

```python
single_col=['all','ent','uncert','diff','ambig','spar']
double_col=['all','diff+ambig','ent+uncert','spar+uncert','spar+diff','ambig+uncert','ent+spar','ambig+spar','ent+diff','diff+uncert','ent+ambig']

single_df=df[single_col]
double_df=df[double_col]
x_ax=df.samples.tolist()
x_ax.insert(0,0)
```

```python
sns.set_theme(style="whitegrid")
sns.set_context("paper", font_scale=1.5)
plot=sns.lineplot(data=single_df, palette="hls", linewidth=1,markers=True, markersize=8)
plot.set_xticklabels(x_ax) 
plot.set(xlabel='Number of samples', ylabel='Average F1-score')
plot.set_ylim(0.60, 0.95) 
plot.spines['top'].set_visible(False)
plot.spines['right'].set_visible(False)
plot.legend(loc='lower right', fontsize=10, frameon=True)
```

```python
sns.set_theme(style="whitegrid")
sns.set_context("paper", font_scale=1.5)
plot=sns.lineplot(data=double_df, palette="hls", linewidth=1,markers=True, markersize=8)
plot.set_xticklabels(x_ax) 
plot.set(xlabel='Number of samples', ylabel='Average F1-score')
plot.set_ylim(0.60, 0.95) 
plot.spines['top'].set_visible(False)
plot.spines['right'].set_visible(False)
plot.legend(loc='lower right', fontsize=10, frameon=True)
```

```python
# 폰트 및 스타일 설정
plt.style.use('seaborn-whitegrid')
sns.set_context("paper", font_scale=1.5)

# 데이터 준비
samples = df['samples']
x = np.arange(len(samples))
methods = df.columns[1:]  # 첫 번째 컬럼 제외

# 그래프 설정
fig, ax = plt.subplots(figsize=(10, 6))

for method in methods:
    ax.plot(x, df[method], marker='o', label=method)

# 축 및 제목 설정
ax.set_xticks(x)
ax.set_xticklabels(samples, rotation=45, ha='right')
ax.set_xlabel("Sample Size")
ax.set_ylabel("Accuracy")
ax.set_title("Comparison of Methods")
ax.legend(loc='lower right', fontsize=10, frameon=True)

# 논문 스타일로 저장
plt.tight_layout()
plt.savefig("experiment_results.png", dpi=300, bbox_inches='tight')
plt.show()

```
