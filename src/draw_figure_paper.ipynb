```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_excel("D:/hyr69/문서/논문/My paper/실험결과.xlsx",sheet_name='Sheet1')
df.head()
```

```python
single_col=['all','ent','uncert','diff','ambig','spar']
double_col=['all','diff+ambig','ent+uncert','spar+uncert','spar+diff','ambig+uncert','ent+spar','ambig+spar','ent+diff','diff+uncert','ent+ambig']

single_df=df[single_col].iloc[:4]
double_df=df[double_col].iloc[:4]

single_df
```

```python
def save_fig(df,save_path):
    x_ax=[50,100,150,200] # x축 레이블
    sns.set_theme(style="whitegrid")
    sns.set_context("paper", font_scale=1.5)
    plot=sns.lineplot(data=df, palette="hls", linewidth=1,markers=True, markersize=8)
    plot.set_xticklabels(x_ax) 
    plot.set_xticks(range(len(x_ax)))
	
    plot.set(xlabel='Number of samples', ylabel='Average F1-score')
    plot.set_ylim(0.60, 0.95)  
    plot.spines['top'].set_visible(False)
    plot.spines['right'].set_visible(False)
    plot.legend(loc='lower right', fontsize=10, frameon=True)
    plt.savefig(save_path, dpi=300)
    plt.show()
    print(f'save {save_path}')
    return 0
```

```python
single_save_path='E:/glass_git/ML-DL/Paper/src/single.png'
double_save_path='E:/glass_git/ML-DL/Paper/src/double.png'

save_fig(single_df,single_save_path)    
save_fig(double_df,double_save_path)    
```
